To have a better idea about the statistical difference between the groups, an F-score is computed that will check if the feature re78 allows to diffenrentiate classes (by computing correlation between the values and the labels) at a level of 0.05 significance.

````python
from sklearn.feature_selection import f_regression

def fisher_result(df):
    
    ns = len(df.index)
    X, y = df.re78.values, df.treat.values
    scores, pvalues = f_regression(X.reshape([ns,1]),y.reshape(ns),center=True)
    
    # convert to one tailed
    mean_diff = np.mean(df[df['treat']==1]['re78'])-np.mean(df[df['treat']==0]['re78'])#subjects-controls 
    pv = float(pvalues)/2
    
    print('average subjects - average controls for re78: {}'.format(mean_diff))
    print('The Fisher test returns a p-value of {} for re78'.format(pv))
````

For cathegorical features analysis, the proportions are computed and displayed in piecharts.

````python
'''Functions to compute the desired proportions'''

# all the functrions return the proportions in a way that can be used for a piechart (i.e prob and 1-prob)

def race_prop(df):
    N = len(df);
    prop1 = np.sum(df['black'])/N;
    prop2 = np.sum(df['hispan'])/N;
    prop0 = 1-(prop1+prop2);
    return [prop0,prop1,prop2]

def married_prop(df):
    married = np.sum(df['married'])/len(df)
    return [married, 1-married]

def nodegree_prop(df):
    nodegree= np.sum(df['nodegree'])/len(df)
    return [nodegree,1-nodegree]
````

````python
'''Define a function for porportions display'''

def cathegories_props(subjects, controls):
    
    # race proportion
    race_props = pd.DataFrame(index = ['white','black','hispan'])
    
    race_props['Subjects - race prop'] = race_prop(subjects)
    race_props['Controls - race prop'] = race_prop(controls)
    
    # married proportion
    married_props = pd.DataFrame(index = ['married','single'])
    
    married_props['Subjects - married prop'] = married_prop(subjects)
    married_props['Controls - married prop'] = married_prop(controls)
    
    # no degree proportions
    nodeg_props = pd.DataFrame(index = ['nodegree','degree'])

    nodeg_props['Subjects - nodegree prop'] = nodegree_prop(subjects)
    nodeg_props['Controls - nodegree prop'] = nodegree_prop(controls)
    
    plt.figure(1)
    race_props.plot.pie(autopct='%.2f',figsize=(8.5,4),subplots=True)
    plt.figure(2)
    married_props.plot.pie(autopct='%.2f',figsize=(8.5,4),subplots=True)
    plt.figure(3)
    nodeg_props.plot.pie(autopct='%.2f',figsize=(8.5,4),subplots=True)
````

*** Matching the groups using networks***
<font color=royalblue size=2>
<b>
By exploring the documentation of this library we found that matching functions allow to solve a maximum matching problem in bipartite graphs. Bipartite graphs are appropriate for our problem, since we have two groups. The maximum matching allows to reduce weighted connexions to the ones that maximize the sum of weights all keeping only paired connexions from a group to the other.

Here to answer the question, we consider the subjects as the limitting group of nodes and controls as the other group of nodes. Maximization of sum of weights should be equivalent to minimizing the sum of absolute differences of propensity scores. This is simply done by defining weights as MAX - abs(diff) where the MAX is the maximum absolute difference computed all over the data so that wieghts are always positive. The derivative of sum of weights then reduces to our matching optimization problem.  


Procedure:

1) Compute an adjacency matrix which contains absolute differences of scores for each node

2) Building the graph by defining nodes as subjects and controls and weights of connexions as the MAX-abs(diff) thanks to adjacency matrix

3) Calling the matching function to build the map

4) Extracting the filtered controls thank to the previous mapping

5) Comparing features again then to assess If the matching worked properly.

````python
import networkx as nx

def match_two_groups(val_subjects, val_controls):
    
    # build the adjacency matrix which defines the weight of connexion between two nodes (subjects/controls)
    N1 = len(val_subjects)
    N2 = len(val_controls)
    A= np.zeros([N1,N2])

    for i in range(0,N1):
        A[i] = np.abs(val_controls-val_subjects[i])
    
    #build the bipartite graph for matching
    B = nx.Graph()
    B.add_nodes_from(np.arange(0,N1), bipartite=0) # Nodes for subjects are indexed from 0 to Nsubjects
    B.add_nodes_from(np.arange(N1,N1+N2), bipartite=1) # Nodes for controls are indexed from Nsubjects+1 to Ncontrols

    maxx = float(np.max(A)) #the max is used to adapt the optimization problem

    for i in range(0,N1):
        for j in range(0,N2):
            B.add_edge(i,j+N1,weight=maxx-A[i][j]) #add the weighted connexions
        
    #perform the matching with networks: this step takes time to run !    
    mapp = nx.max_weight_matching(B)
    
    # reconstruct the data set
    subjects_ids = np.arange(0,N1)
    controls_ids = []

    for i in range(0,N1):
        num = mapp[i] #find the index matched (control) to the current index (subject)
        controls_ids.append(num-N1) #adapt the index for the control df index
        
        
    return controls_ids
````

# Improve matching
<font color=royalblue size=2>
<b>
In the previous question, we found that our matching was pretty good. Except for the age repartition and ages. The problem is that age repartitions were more similar before the matching meaning that to match other features like races, re74 etc, the algorithm had to slightly unmatch ages. 

But note that the problem would be simply solved by adding age absolute difference as a cost when building the Adjacency matrix (see "match_to_groups" function).

For this study, we made the hypothesis that the races may play a most important role in the propensity to receive the treatment. Probably because of discrimination making it more difficult to non-white people to get a good job.
Therefore we try to improve the matching by ensuring a more equal proportion of non-white people in both groups or equivalently a more ballanced proportion of white people.

For this we will:

1) Build race groups

2) Match subjects from a group only with controls of the same group, i.e black and black, white and white and hispan and hispan.

3) Unmatched subjects are matched based on propensity scores with unmatched controls.

By this way we put the race as a priority for matching and then, take other features into account.

````python
'''Implicitely categorize the races'''

def build_races(df):
    new_df = df.copy()
    race = np.array(1*df['black']+2*df['hispan']) #encode the race with a given coefficient 
    new_df['race'] = race
    new_df['memm_idx'] = np.arange(0,len(df.index)) #this step memmorizes the initial index of a subject in the df
    # this will be helpful for other steps
    
    black = new_df[new_df['race']==1]
    hispan = new_df[new_df['race']==2]
    white = new_df[new_df['race']==0]
    
    return black, hispan, white, race #retain the list with races to extract propensity scores later
````

<font color=royalblue size=2>
<b>
Now we would like to check the limiting sizes of groups of races comparing subjects and controls.

````python
black_subjects, hispan_subjects, white_subjects, race_subjs = build_races(subjects)

len(black_subjects.index), len(hispan_subjects.index), len(white_subjects.index)
````

````python
black_controls, hispan_controls, white_controls, race_ctrls = build_races(controls)

len(black_controls.index), len(hispan_controls.index), len(white_controls.index)
````

Then the black subjects unmatched will be compared to the rest of the controls.

````python
'''Here we match hispans with hispans and whites with whites'''

hispan_match = match_two_groups(propensity_scores_subjects[np.where(race_subjs==2)],
                                propensity_scores_controls[np.where(race_ctrls==2)])

white_match = match_two_groups(propensity_scores_subjects[np.where(race_subjs==0)],
                               propensity_scores_controls[np.where(race_ctrls==0)])
````

````python
'''Now, make the inverse match of black controls with black subjects'''

black_inverse_match = match_two_groups(propensity_scores_controls[np.where(race_ctrls==1)],
                                       propensity_scores_subjects[np.where(race_subjs==1)])
````

````python
'''Make the matching between the unretained subjects and unretained controls'''

# build the unmathed dataframes for controls in each specific group
hisp_no_ids, hisp_no_retained = no_retained_ids(hispan_controls, hispan_match)
white_no_ids, white_no_retained  = no_retained_ids(white_controls, white_match)

# build the unmathed index list for controls thanks to the memmorized initial indexes
no_retained_ctrls = pd.concat([hisp_no_retained,white_no_retained])
no_retained_ctrls_ids = list(no_retained_ctrls['memm_idx'])

# the same previous steps are repeated to build the no retained subjects indices
black_subjs_no_ids, black_subjs_no_retained = no_retained_ids(black_subjects, black_inverse_match)
no_retained_subjs_ids = list(black_subjs_no_retained['memm_idx'])
````

````python
'''Make the last group matching'''

remain_match = match_two_groups(propensity_scores_subjects[no_retained_subjs_ids],
                                propensity_scores_controls[no_retained_ctrls_ids])
````

````python
'''Re-build the controls group by concatenating all the retained subgroups'''

new_controls_hisp = retained_df(hispan_controls, hispan_match)
new_controls_white = retained_df(white_controls, white_match)
new_controls_remain = retained_df(no_retained_ctrls, remain_match)
new_controls_black = black_controls

retained_df_controls = pd.concat([new_controls_hisp,new_controls_white,new_controls_black,new_controls_remain])
````

c. Then, split your dataset into a training, a testing and a validation set (10% for validation and 10% for testing). Each observation should be paired with its corresponding label (the article category).</b>

````python
def split_dataset(data, labels, train_proportion=0.8, validation_proportion=0.1, test_proportion=0.1, seed=1):
    num_datapoints = data.shape[0]
    # Generate a random permutation of the rows' indices
    np.random.seed(seed)
    indices = np.random.permutation(num_datapoints)
    # Compute indices to sample the random index with the correct number of values for each set
    nb_tr_data = int(np.ceil(num_datapoints*train_proportion))
    nb_validation_data = int(np.ceil(num_datapoints*validation_proportion))
    nb_te_data = int(np.ceil(num_datapoints*test_proportion))
    #  Make up for rounding errors
    if(num_datapoints < nb_tr_data + nb_validation_data + nb_te_data) :
        nb_tr_data -= (nb_tr_data + nb_validation_data + nb_te_data) - num_datapoints
    assert(nb_tr_data + nb_validation_data + nb_te_data <= num_datapoints)
    # Sample data and labels
    tr_data = data[:nb_tr_data]
    tr_labels = labels[:nb_tr_data]
    v_data = data[nb_tr_data : nb_tr_data + nb_validation_data]
    v_labels = labels[nb_tr_data : nb_tr_data + nb_validation_data]
    te_data = data[nb_tr_data + nb_validation_data : ]
    te_labels = labels[nb_tr_data + nb_validation_data : ]
    return tr_data, tr_labels, v_data, v_labels, te_data, te_labels
````

Try to fine-tune the parameters of your predictor on your validation set using a simple grid search on the number of estimator "n_estimators" and the max depth of the trees "max_depth". 



````python
## Let's define here all necessary functions to train ou model 
##

# Model performance measurement
def compute_accuracy(y1, y2):
    return np.mean(y1 == y2)


# Grid search
def grid_search(tr_data, tr_labels, v_data, v_labels, range_estimators, range_depth) : 
    accuracy_accumulator = []
    for estimator_itr, n_estimators in enumerate(range_estimators) :
        #print("n_estimators ", estimator_itr+1," in ", len(range_estimators), " : ", n_estimators)
        accuracy_accumulator.append([])
        for depth_itr, max_depth in enumerate(range_depth) : 
            #print("    | max_depth ", depth_itr+1," in ", len(range_depth), " : ", max_depth, end='')
            rfc = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, oob_score=True, random_state=0)
            rfc.fit(tr_data, tr_labels)
            predictions = np.array(rfc.predict(v_data))
            accuracy = compute_accuracy(predictions, v_labels)
            #print("    --- accuracy : ", accuracy)
            accuracy_accumulator[estimator_itr].append(accuracy)
        if estimator_itr%2==0 : print((estimator_itr+1) / len(range_estimators), "% completed")
    return np.array(accuracy_accumulator)


# Print out results as a heatmap
def print_gridsearch_results(accuracy_accumulator, range_estimators, range_depth) : 
    fig, axis = plt.subplots()
    heatmap = axis.pcolor(accuracy_accumulator.transpose(), cmap=plt.cm.Blues)

    axis.set_xticks(np.arange(accuracy_accumulator.shape[0])+0.5, minor=False)
    axis.set_xticklabels(range_estimators, minor=False)
    axis.set_yticks(np.arange(accuracy_accumulator.shape[1])+0.5, minor=False)
    axis.set_yticklabels(range_depth, minor=False)

    plt.colorbar(heatmap)
    plt.ylabel("max_depth")
    plt.xlabel("n_estimators")

    plt.show()
````

````python
## Parameters 
##
range_estimators = np.linspace(300, 1500, num=13, endpoint=True, dtype=np.int64)
range_depth = np.logspace(4, 8, base=2, num=5, endpoint=True, dtype=np.int64)

accuracy_accumulator = grid_search(t_data, t_labels, v_data, v_labels, range_estimators, range_depth)
````

Now, let's define a function to visualize our confusion matrix : 

````python
""" This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    Credits to sklearn documentation :
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
"""
def plot_confusion_matrix(cm, classes_cat1, 
                          classes_cat2=None,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.yticks(np.arange(len(classes_cat1)), classes_cat1)
    if not (classes_cat2 is None) : plt.xticks(np.arange(len(classes_cat1)), classes_cat2, rotation=90)
    else : plt.xticks(np.arange(len(classes_cat1)), classes_cat1, rotation=90)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Predicted label')
    plt.xlabel('True label')
    plt.show()
````

It's more convenient to visualize the results by first grouping them by category

````python
import itertools as itertools


''' Group the results by category
'''
def groupby_cats():
    res = []
    for cat1 in categories : 
        for cat2 in categories : 
            res.append(get_cat_intersection(cat1, cat2, sum_=True))
    return np.array(res, dtype=np.int64).reshape((len(categories), len(categories)))


''' Returns a subcategory of the confusion matrix
'''
def get_cat_intersection(cat1, cat2, sum_=False) : 
    labels1 = get_cat_labels(cat1)
    labels2 = get_cat_labels(cat2)
    misclassified_as_cat1 = cm_global_df[labels2].loc[labels1]
    if sum_ : 
        return misclassified_as_cat1.sum(axis=1).sum(axis=0)
    else : 
        return np.array(misclassified_as_cat1.as_matrix(), dtype=np.int64)
    

''' Helper : given a category name, returns all label names in it
'''
def get_cat_labels(cat_name) : 
    wanted_mask = labels.map(lambda e : cat_name == e[:len(cat_name)])
    wanted_labels = labels[wanted_mask]
    return wanted_labels

    
''' Returns a subcategory of the confusion matrix
    Wraps 'get_cat_intersection'
'''
def get_cat(cat_name) : 
    return get_cat_intersection(cat_name, cat_name)
````

````python
plot_confusion_matrix(groupby_cats(), categories, title='Confusion matrix across categories')
````

````python
''' Plots the confusion matrix of mislabelling as cat1 elements of cat2
    (labelled=cat2, actual=cat1)
'''
def plot_inter_cm(cat1, cat2, normalize=True) : 
    inter_cm = get_cat_intersection(cat1, cat2, sum_=False)
    cat1_labels = get_cat_labels(cat1)
    cat2_labels = get_cat_labels(cat2)
    plot_confusion_matrix(inter_cm, 
                          cat1_labels, 
                          classes_cat2=cat2_labels,
                          normalize=normalize,
                          title="Inter-Confusion Matrix for categories "+cat1+" and "+cat2)
````
